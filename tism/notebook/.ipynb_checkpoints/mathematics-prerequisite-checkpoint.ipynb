## 向量
在進行統計建模時，一筆觀測值（observation）常透過向量（vector）來表徵。令 $v$ 表示一 $P$ 維之行向量（column vector），則其包含了 $v_1,v_2,...,v_P$ 共 $P$ 個元素（element），按順序由上至下排列而成，即

$$
v =
\begin{pmatrix}
v_1 \\
v_2 \\
\vdots \\
v_P
\end{pmatrix}
$$

在此講義中，我們將使用 $v_p$ 來表示 $v$ 的第 $p$ 個元素。


```python
import numpy as np
import matplotlib.pyplot as plt
plt.ion()

data = np.random.randn(2, 100)
fig, ax = plt.subplots()
ax.scatter(*data, c=data[1], s=100*np.abs(data[0]));
```
